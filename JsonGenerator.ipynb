## JSON Generator

```
# Importing the libraries
import pandas as pd
import json
```

```
# Defining custom parameters
root_dir = '/content/drive/My Drive/Datasets'
```

```
# Getting the sheet names
xlsxfile = pd.ExcelFile(f'{root_dir}/Intents.xlsx')
sheetslist = xlsxfile.sheet_names
```

```
# Creating the json file
dictlist = []
i = 0
for sheet in sheetslist:
  temp_df = pd.read_excel(f'{root_dir}/Intents.xlsx', sheet_name = sheet)
  temp_df_p = temp_df['Pattern'].tolist()
  temp_df_r = [string for string in temp_df['Response'] if not pd.isnull(string)]
  if i == 0:
    tempdic = {"tag": str(sheet),
               "patterns": temp_df_p,
               "responses": temp_df_r,
               "context_set": ""
               }
    dictlist.append(tempdic)
  else:
    tempdic = {"tag": str(sheet),
               "patterns": temp_df_p,
               "responses": temp_df_r,
               "context_set": ""
               }
    dictlist.append(tempdic)    
  i += 1
maindict = {"intents": dictlist}
jsonfile = open(f'{root_dir}/intents.json', "w") 
json.dump(maindict, jsonfile, indent = 4) 
jsonfile.close() 
```
