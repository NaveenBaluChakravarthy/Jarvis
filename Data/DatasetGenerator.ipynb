```
# Importing the libraries
import json
import pandas as pd
from itertools import product
from google.colab import files
```

```
# Helper Functions
class Helper():
    def __init__(self):
        pass

    def finish(self, q, a):
        questions.append(q)
        answers.append(a)

    def get_data(self):
        players = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Players')
        aliases = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Aliases')
        music = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Music')
        locations = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Locations')
        dialogues = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Dialogues')
        happenings = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Happenings')
        misc = pd.read_excel(f'{C_Root}/{C_Movie}.xlsx', sheet_name = 'Misc')
        return players, aliases, music, locations, dialogues, happenings, misc  

    def build_references(self):
        d_aliases = {'Stanlee' : ['Father of MCU']}
        d_article = {'A': 'is'}
        d_legalname = {'A': 'Alpha'}
        for i in range(len(aliases)):
            name = aliases['Common Name'][i]
            aka = str(aliases['AKA'][i]).split(', ')
            if aliases['Gender'][i] in ['Masculine', 'Feminine', 'Neuter']:
                article = 'is'
            elif aliases['Gender'][i] in ['Group']:
                article = 'are'
            else:
                print("Not a valid gender!!")
            if aka == ['nan']:
                aka = []
            d_aliases[name] = aka
            d_article[name] = article
            d_legalname[name] = aliases['Formal Name'][i]

        return d_aliases, d_article, d_legalname, [], []

    def qa_players(self):
        for i in range(len(players)):
            p1 = players['Movie'][i]
            p2 = players['Char A'][i]
            p3 = players['Relation'][i]
            p4 = players['Char B'][i]
            p3 = p3.split('/')
            try:
                chara_art = d_article[p2]
            except KeyError:
                chara_art = 'is'
            try:
                charb_art = d_article[p4]
            except KeyError:
                charb_art = 'is'
            try:
                p2 = d_aliases[p2] + [p2] + [d_legalname[p2]]
            except KeyError:
                p2 = [p2]
            try:
                p4 = d_aliases[p4] + [p4] + [d_legalname[p4]]
            except KeyError:
                p4 = [p4]
            for chara in p2:
                for charb in p4:
                    q = f"In the movie {p1}, who {chara_art} the {p3[0]} of {charb}?"
                    a = f"In the movie {p1}, {chara} {chara_art} the {p3[0]} of {charb}."
                    self.finish(q, a)
                    q = f"Who {chara_art} the {p3[0]} of {charb}?"
                    a = f"{chara} {chara_art} the {p3[0]} of {charb}."
                    self.finish(q, a)
                    if len(p3) > 1:
                        q = f"In the movie {p1}, who {charb_art} the {p3[1]} of {chara}?"
                        a = f"In the movie {p1}, {charb} {charb_art} the {p3[1]} of {chara}."
                        self.finish(q, a)
                        q = f"Who {charb_art} the {p3[1]} of {chara}?"
                        a = f"{charb} {charb_art} the {p3[1]} of {chara}."
                        self.finish(q, a)            

    def qa_aliases(self):
        for i in range(len(aliases)):
            p1 = aliases['Common Name'][i]
            p2 = aliases['Formal Name'][i]
            p3 = aliases['Nickname'][i]
            p4 = aliases['Persona'][i]
            p5 = aliases['Actor'][i]
            p6 = aliases['Movie'][i]
            p7 = aliases['Gender'][i]
            p8 = aliases['AKA'][i]

            try:
                chara_art = d_article[p1]
            except KeyError:
                chara_art = 'is'
            try:
                p1 = d_aliases[p1] + [p1]
            except KeyError:
                p1 = [p1]

            for chara in p1:
                q = f"In the movie {p6}, who {chara_art} {chara}?"
                a = f"In the movie {p6}, {chara} {chara_art} a {p4}."
                self.finish(q, a)
                q = f"Who {chara_art} {chara}?"
                a = f"{chara} {chara_art} a {p4}."
                self.finish(q, a)  
                if type(p5) == str:
                    q = f"Who portrayed {chara}?"
                    a = f"{p5} portrayed {chara}."
                    self.finish(q, a)  
                    q = f"In the movie {p6}, who portrayed {chara}?"
                    a = f"In the movie {p6}, {p5} portrayed {chara}."
                    self.finish(q, a) 
                if type(p2) == str:
                    q = f"What is the formal name of {chara}?"
                    a = f"{p2} is the formal name of {chara}."
                    self.finish(q, a)  
                    q = f"In the movie {p6}, what is the formal name of {chara}?"
                    a = f"In the movie {p6}, {p2} is the formal name of {chara}."
                    self.finish(q, a) 
                if type(p3) == str:
                    x1 = p3.split(', ')
                    if len(x1) > 1:
                        x2 = 'are the nicknames of'
                    else:
                        x2 = 'is the nickname of'
                    q = f"In the movie {p6}, what {x2} {chara}?"
                    a = f"In the movie {p6}, {p3} {x2} {chara}."
                    self.finish(q, a)
                    q = f"What {x2} {chara}?"
                    a = f"{p3} {x2} {chara}."
                    self.finish(q, a)
                    q = f"In the movie {p6}, who gave the nickname to {chara}?"
                    a = f"In the movie {p6}, Tony gave the nickname to {chara}."
                    self.finish(q, a)
                    q = f"Who gave the nickname to {chara}?"
                    a = f"Tony gave the nickname to {chara}."
                    self.finish(q, a)
                if type(p8) == str:
                    x1 = p8.split(', ')
                    if len(x1) > 1:
                        x2 = 'are the aliases of'
                    else:
                        x2 = 'is the alias of'
                    q = f"In the movie {p6}, what {x2} {chara}?"
                    a = f"In the movie {p6}, {p8} {x2} {chara}."
                    self.finish(q, a)
                    q = f"What {x2} {chara}?"
                    a = f"{p8} {x2} {chara}."
                    self.finish(q, a)

                q = f"In the movie {p6}, who {chara_art} {p2}?"
                a = f"In the movie {p6}, {p2} {chara_art} a {p4}."
                self.finish(q, a)
                q = f"Who {chara_art} {p2}?"
                a = f"{p2} {chara_art} a {p4}."
                self.finish(q, a)            
            
    def qa_music(self):
        for i in range(len(music)):
            p1 = music['Movie'][i]
            p2 = music['Scene'][i]
            p3 = music['Soundtrack'][i]
            q = f"In the movie {p1}, what soundtrack plays in the {p2} scene?"
            a = f"In the movie {p1}, {p3} soundtrack plays in the {p2} scene."
            self.finish(q, a)
            q = f"What soundtrack plays in the {p2} scene?"
            a = f"{p3} soundtrack plays in the {p2} scene."
            self.finish(q, a)

    def qa_locations(self):
        for i in range(len(locations)):
            p1 = locations['Movie'][i]
            p2 = locations['Scene'][i]
            p3 = locations['Location'][i]
            p4 = locations['Type'][i]
            if p4 == 'Event':
                x1 = f'does the {p2} happen'
                x2 = f'{p2} happens at {p3}'
            elif p4 == 'Place':
                x1 = f'is the {p2} located'
                x2 = f'{p2} is located in {p3}'
            else:
                print('Bad event type!!')
                x1, x2 = 'A', 'A'
            if x1+x2 != 'AA':
                q = f"In the movie {p1}, where {x1}?"
                a = f"In the movie {p1}, {x2}."
                self.finish(q, a)
                q = f"where {x1}?"
                a = f"{x2}."
                self.finish(q, a)

    def qa_dialogues(self):
        for i in range(len(dialogues)):
            p1 = dialogues['Movie'][i]
            p2 = dialogues['Scene'][i]
            p3 = dialogues['Speaker'][i]
            p4 = dialogues['Receiver'][i]
            p5 = dialogues['Context'][i]
            p6 = dialogues['Dialogue'][i]

            if type(p5) == str:
                x1 = f" about {p5}"
            else:
                x1 = ''
            try:
                p3 = d_aliases[p3] + [p3] + [d_legalname[p3]]
            except KeyError:
                p3 = [p3]
            try:
                p4 = d_aliases[p4] + [p4] + [d_legalname[p4]]
            except KeyError:
                p4 = [p4]

            for chara in p3:
                for charb in p4:
                    q = f"In the movie {p1}, in the {p2} scene, what did {chara} say to {charb}{x1}?"
                    a = f"{p6}"
                    self.finish(q, a)
                    q = f"In the {p2} scene, what did {chara} say to {charb}{x1}?"
                    a = f"{p6}"
                    self.finish(q, a)

    def qa_happenings(self):
        for i in range(len(happenings)):
            p1 = happenings['Movie'][i]
            p2 = happenings['Character'][i]
            p3 = happenings['Event'][i]
            p4 = happenings['Time'][i]

            repl_dict = {}
            i = 0
            for key in d_aliases.keys():
                if key in p2:
                    p2 = p2.replace(key, f"zzzz{str(i)}zzzz")
                    try:
                        value = d_aliases[key] + [key] + [d_legalname[key]]
                    except KeyError:
                        value = [key]
                    repl_dict[f"zzzz{str(i)}zzzz"] = value
                    i += 1
            chars = len(repl_dict)
            charlist = [repl_dict[f"zzzz{str(j)}zzzz"] for j in range(chars)] 
            eval_string = "product(" + ', '.join([f'charlist[{k}]' for k in range(chars)]) + ")"
            for combos in eval(eval_string):
                px = p2
                for m in range(len(combos)):
                    px = px.replace(f"zzzz{str(m)}zzzz", combos[m])
                q = f"In the movie {p1}, when did {px} {p3}?"
                a = f"{p4}."
                self.finish(q, a) 
                q = f"When did {px} {p3}?"
                a = f"{p4}."
                self.finish(q, a) 

    def qa_misc(self):
        for i in range(len(misc)):
            p1 = misc['Movie'][i]
            p2 = misc['Question'][i]
            p3 = misc['Answer'][i]

            repl_dict = {}
            i = 0
            for key in d_aliases.keys():
                if key in p2:
                    p2 = p2.replace(key, f"zzzz{str(i)}zzzz")
                    try:
                        value = d_aliases[key] + [key] + [d_legalname[key]]
                    except KeyError:
                        value = [key]
                    repl_dict[f"zzzz{str(i)}zzzz"] = value
                    i += 1
            chars = len(repl_dict)
            charlist = [repl_dict[f"zzzz{str(j)}zzzz"] for j in range(chars)] 
            eval_string = "product(" + ', '.join([f'charlist[{k}]' for k in range(chars)]) + ")"
            for combos in eval(eval_string):
                px = p2
                for m in range(len(combos)):
                    px = px.replace(f"zzzz{str(m)}zzzz", combos[m])
                q = f"{px}"
                a = f"{p3}"
                self.finish(q, a) 
```

```
X = Helper()
players, aliases, music, locations, dialogues, happenings, misc = X.get_data()
d_aliases, d_article, d_legalname, questions, answers = X.build_references()
X.qa_players()
X.qa_aliases()
X.qa_music()
X.qa_locations()
X.qa_dialogues()
X.qa_happenings()
X.qa_misc()
dataset = pd.DataFrame({'Questions': questions, 'Answers': answers})
dataset.to_csv(f'{C_Root}/MarvelCinematicUniverse_{C_Movie}.csv', index = False)
files.download(f'{C_Root}/MarvelCinematicUniverse_{C_Movie}.csv')
```
